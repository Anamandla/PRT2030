```python
import pandas as pd
import matplotlib.pyplot as plt
gold = pd.read_csv('gold.csv')
gold.info()
```

```python
gold['Date'] = pd.to_datetime(gold['Date'])
gold.head()
```

```python
gold.plot(x= 'Date', y= 'Price', kind= 'line')
plt.title('Gold Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')

plt.show()
```

```python
gold.set_index('Date', inplace=True)
annual_periods = pd.date_range(start='1950-01-01', end='2020-01-01', freq='YS')
print(annual_periods)
```

```python
gold_reindexed = gold.reindex(annual_periods)
gold_reindexed.head()
```

```python
gold_reindexed.plot(y='Price', kind='line')
plt.title("Annual Gold Prices")
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
gold_quarterly = gold.resample('Q').mean()
gold_quarterly.plot(y='Price', kind='line')
plt.title('Quarterly Gold Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
gold_yearly = gold.resample('Y').mean()
gold_yearly.plot(y='Price', kind='line')
plt.title('Yearly Gold Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
gold['Rolling Mean'] = gold['Price'].rolling(window=12).mean()
gold.plot(y='Rolling Mean', kind='line')
plt.title('Rolling Mean of Gold Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
